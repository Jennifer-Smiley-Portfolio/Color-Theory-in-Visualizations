```python
bars = alt.Chart(df).mark_bar(color = "darkorange").encode(
    x='count(Orange)',
    y=alt.Y("Orange", sort='-x'),
)
text = bars.mark_text(
    align='left',
    baseline='middle',
    dx=3   # Nudges text to right so it doesn't appear on top of the bar
).encode(
    text='count(Orange)'
)
(bars + text).properties()
```
